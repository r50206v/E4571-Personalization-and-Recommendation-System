```python
def checkID(series):
    tmp = series.value_counts()
    print("unique ID: ", tmp.loc[tmp == 1].shape[0])
    print("duplicates ID: ", tmp.loc[tmp != 1].shape[0])
```

```python
checkID(user_id['user_id'])
```

```python
checkID(business_id['business_id'])
```

```python
df = pd.read_csv('yelp_dataset/review.csv')
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d %H:%M:%S')
df.head()
```

```python
print(df.shape)
df.describe()
```

```python
df.dtypes
```

```python
rr = df.loc[
    df['business_id'].isin(business_id['business_id'].values.tolist()) &\
    df['user_id'].isin(user_id['user_id'].values.tolist())
]
```

```python
print(rr.shape)
rr.dtypes
```

```python
rr.head()
```

```python
date_after = datetime.strptime('2015-01-01', "%Y-%m-%d")
date_after
```

```python
result = rr.loc[rr['date'] >= date_after]
result['date'].sort_values().iloc[[0,1,2,-3,-2,-1]]
```

```python
result.dtypes
```

```python
result.to_csv("yelp_dataset/review_cleaned.csv", index=False)
```
